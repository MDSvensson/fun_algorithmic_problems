# Simplify Path

```python
class Solution:
    def simplify_path(self, path: str) -> str:
        cur = ""
        stack = []
        for c in path:
            if c == "/":
                if cur == "..":
                    if stack:
                        stack.pop()
                elif cur != "." and cur != "":
                    stack.append(cur)
                cur = ""
            else:
                cur += c
        return "/" + "/".join(stack)


sol = Solution()
# sol.simplify_path("/../abc//./def/")
# sol.simplify_path("/home/")
sol.simplify_path(
    "/home/user/Documents/../Pictures"
)  # why we need the slash at the end
```

# Generate Parentheses

```python
from typing import List


class Solution:
    def generate_parenthesis(self, n: int) -> List[str] | None:
        res = []
        stack = []

        # only add open parenthesis if n_open < n
        # only add closing parenthesis if n_closed < n_open
        # valid (base case) IIF n_open == n_closed
        def backtracking(n_open, n_closed):
            # base case
            if n == n_open == n_closed:
                res.append("".join(stack))
                return
            if n_open < n:
                stack.append("(")
                backtracking(n_open + 1, n_closed)
                stack.pop()
            if n_closed < n_open:
                stack.append(")")
                backtracking(n_open, n_closed + 1)
                stack.pop()

        backtracking(0, 0)
        return res


sol = Solution()
sol.generate_parenthesis(3)
```
