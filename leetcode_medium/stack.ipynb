# Simplify Path

```python
class Solution:
    def simplify_path(self, path: str) -> str:
        cur = ""
        stack = []
        for c in path:
            if c == "/":
                if cur == "..":
                    if stack:
                        stack.pop()
                elif cur != "." and cur != "":
                    stack.append(cur)
                cur = ""
            else:
                cur += c
        return "/" + "/".join(stack)


sol = Solution()
# sol.simplify_path("/../abc//./def/")
# sol.simplify_path("/home/")
sol.simplify_path(
    "/home/user/Documents/../Pictures"
)  # why we need the slash at the end
```
