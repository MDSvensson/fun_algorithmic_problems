# Vacation Scheduler

```python
from typing import List


class Solution:
    def vac_len(self, schd: List[str], pto: int) -> int:
        l = 0
        res_len = 0
        count = {}
        for r in range(len(schd)):
            count[schd[r]] = 1 + count.get(schd[r], 0)
            while (r - l + 1) - count.get("H", 0) > pto:
                count[schd[l]] -= 1
                l += 1
            res_len = max(res_len, r - l + 1)
        return res_len


sol = Solution()
sol.vac_len(["W", "H", "W", "W", "W", "H", "H", "H", "W"], 3)
```

# Best Time to Buy and Sell Stock

```python
from typing import List


class Solution:
    def max_profit(self, s: List) -> int:
        l = 0
        cur_profit = 0
        for r in range(1, len(s)):
            if s[r] <= s[l]:
                # update l to be = to r
                l = r
            cur_profit = max(cur_profit, s[r] - s[l])
        return cur_profit


sol = Solution()
sol.max_profit([7, 1, 5, 3, 6, 4])
```
